## Simple English to Chinese Translation Function

````python
#translates English to Chinese for the initial search of wikipedia in language being learned
def english2chinese(topic, display=False):
    """ Translates a word or words from English into Traditional Mandarin"""
    # Instantiates a client
    translate_client = translate.Client()
    text = topic
    # Defines the target language
    target = 'zh-TW'
    # Translates some text into traditional Chinese
    translation = translate_client.translate(text, target_language=target)
    text_ = translation['translatedText']
    if display == True:
        print(text_)
        return text_
    else:
        return text_
````

## Simple Chinese to English Translation Function

````python
# translates Chinese to English
def chinese2english(topic):
    # Instantiates a client
    translate_client = translate.Client()
    text = topic
    # Defines the target language
    target = 'en'
    # Translates some text into traditional Chinese
    translation = translate_client.translate(text, target_language=target)
    text_ = translation['translatedText']
    if display == True:
        print(text_)
        return text_
    else:
        return text_
````

## English To Chinese 

````python
def e2c_say_anything():
    translate_client = translate.Client()
    text = input('What word would you like to say in Chinese(Mandarin)：')
    # Defines the target language
    target = 'zh-TW'
    # Translates some text into traditional Chinese
    translation = translate_client.translate(text, target_language=target)
    text_ = translation['translatedText']
    # Build the voice request, select the language code ("en-US") and the ssml
    # voice gender ("neutral")
    
    #add pinyin
    pinyin_with_numbers = p.get_pinyin(text_, tone_marks='numbers')
    tts = gTTS(text=text_, lang='zh-TW')
    tts.save("good.mp3")
    
    print("Input: {}".format(text))
    print("Traditional Chinese Translation: {}".format(text_))
    print("Pinyin Translation with tone numbers: {}".format(pinyin_with_numbers))
    return os.system("mpg321 good.mp3")
````

# Wikipedia Search

````python
def search_wikipedia(topic, display=False,results=3):
    wikipedia.set_lang('zh')
    search_phrase = english2chinese(topic)
    pages = wikipedia.search(search_phrase, results=results)
    if display == True:
        print('the pages in native language', pages)
        return pages
    else:
        return pages
````

````python
def get_content(topic):
    """Takes a topic Parameter and returns the wikipedia page that best matches that topic in Chinese"""
    wikipedia.set_lang('zh')
    search_phrase = english2chinese(topic)
    wiki_page = wikipedia.page(search_phrase,preload=True)
    return wiki_page.content
````

````python
def get_top_n_words(corpus, n=None):
    """
    List the top n words in a vocabulary according to occurrence in a text corpus.
    
    get_top_n_words(["I love Python", "Python is a language programming", "Hello world", "I love the world"]) -> 
    [('python', 2),
     ('world', 2),
     ('love', 2),
     ('hello', 1),
     ('is', 1),
     ('programming', 1),
     ('the', 1),
     ('language', 1)]
    """
    
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    total_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, total_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)

    
    return words_freq[:n]
````

Keyword extraction based on TF-IDF algorithm
import jieba.analyse

Jieba.analyse.extract_tags(sentence, topK=20, withWeight=False, allowPOS=())
Sentence is the text to be extracted
topK is the keyword that returns the most weights of TF/IDF. The default value is 20
withWeight is to return the keyword weight value together, the default value is False
allowPOS only includes words with specified part of speech. The default value is null, that is, no filtering.
jieba.analyse.TFIDF(idf_path=None) Create a new TFIDF instance, idf_path is the IDF frequency file
Code example (keyword extraction)

Https://github.com/fxsjy/jieba/blob/master/test/extract_tags.py

The reverse file frequency (IDF) text corpus used for keyword extraction can be switched to the path of a custom corpus

Usage: jieba.analyse.set_idf_path(file_name) # file_name is the path to the custom corpus
Custom corpus example: https://github.com/fxsjy/jieba/blob/master/extra_dict/idf.txt.big
Example usage: https://github.com/fxsjy/jieba/blob/master/test/extract_tags_idfpath.py
The word corpus used in keyword extraction can be switched to the path of the custom corpus.

Usage: jieba.analyse.set_stop_words(file_name) # file_name is the path to the custom corpus
Custom corpus example: https://github.com/fxsjy/jieba/blob/master/extra_dict/stop_words.txt
Example usage: https://github.com/fxsjy/jieba/blob/master/test/extract_tags_stop_words.py
Key words together return keyword weight value example

Example usage: https://github.com/fxsjy/jieba/blob/master/test/extract_tags_with_weight.py
Keyword extraction based on TextRank algorithm
Jieba.analyse.textrank(sentence, topK=20, withWeight=False, allowPOS=('ns', 'n', 'vn', 'v')) Direct use, the interface is the same, pay attention to the default filter part of speech.
jieba.analyse.TextRank() Create a new custom TextRank instance
Algorithm Paper: TextRank: Bringing Order into Texts

Basic ideas:
Fragment the text of the keyword to be extracted
With a fixed window size (default is 5, adjusted by the span attribute), the co-occurrence relationship between words, build a graph
Calculate the PageRank of the node in the graph, pay attention to the undirected weighted graph
Example of use:
See test/demo.py
